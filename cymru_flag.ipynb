```python
import PIL;
import requests;
from io import BytesIO;

def print_1(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+0, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  return 2;

def print_2(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+0, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+1, y+4] = text_colour;
  pixels[x+2, y+4] = text_colour;
  return 4;

def print_3(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+1, y+4] = text_colour;
  pixels[x+2, y+4] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+3] = text_colour;
  return 4;

def print_4(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  return 4;

def print_5(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+1, y+4] = text_colour;
  pixels[x+2, y+4] = text_colour;
  return 4;

def print_6(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+0, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+1, y+4] = text_colour;
  return 4;

def print_7(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  return 4;

def print_8(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+0, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+1, y+2] = text_colour;
  pixels[x+1, y+4] = text_colour;
  return 4;

def print_9(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+1, y+2] = text_colour;
  return 4;

def print_0(x, y, pixels, text_colour):
  pixels[x+0, y+0] = text_colour;
  pixels[x+0, y+1] = text_colour;
  pixels[x+0, y+2] = text_colour;
  pixels[x+0, y+3] = text_colour;
  pixels[x+0, y+4] = text_colour;
  pixels[x+2, y+0] = text_colour;
  pixels[x+2, y+1] = text_colour;
  pixels[x+2, y+2] = text_colour;
  pixels[x+2, y+3] = text_colour;
  pixels[x+2, y+4] = text_colour;
  pixels[x+1, y+0] = text_colour;
  pixels[x+1, y+4] = text_colour;
  return 4;

def print(x, y, n, pixels, text_colour):
  if n == 0:
    return print_0(x, y, pixels, text_colour);
  if n == 1:
    return print_1(x, y, pixels, text_colour);
  if n == 2:
    return print_2(x, y, pixels, text_colour);
  if n == 3:
    return print_3(x, y, pixels, text_colour);
  if n == 4:
    return print_4(x, y, pixels, text_colour);
  if n == 5:
    return print_5(x, y, pixels, text_colour);
  if n == 6:
    return print_6(x, y, pixels, text_colour);
  if n == 7:
    return print_7(x, y, pixels, text_colour);
  if n == 8:
    return print_8(x, y, pixels, text_colour);
  if n == 9:
    return print_9(x, y, pixels, text_colour);
  
  return 0;

colours = [
           (255, 69, 0, 255),
           (255, 168, 0, 255),
           (255, 214, 53, 255),
           (0, 163, 104, 255),
           (126, 237, 86, 255),
           (36, 80, 164, 255),
           (54, 144, 234, 255),
           (81, 233, 244, 255),
           (129, 30, 159, 255),
           (180, 74, 192, 255),
           (255, 153, 170, 255),
           (156, 105, 38, 255),
           (0, 0, 0, 255),
           (137, 141, 141, 255),
           (212, 215, 217, 255),
           (255, 255, 255, 255),
];

def get_best_text_colour(input):
  if input == (0, 0, 0, 255):
    return (127, 127, 127, 255);
  else:
    return (0, 0, 0, 255);

def colour_distance(c1, c2):
  (c1r, c1g, c1b, c1a) = c1;
  (c2r, c2g, c2b, c2a) = c2;
  rmean = (c1r + c2r) / 2;
  r = c1r - c2r;
  g = c1g - c2g;
  b = c1b - c2b;
  return float(
    ((int(((512+rmean)*r*r))>>8) + 4*g*g + (int(((767-rmean)*b*b))>>8))**0.5  
  );


def get_nearest(input):
  least_dist = float('inf');
  output = input;
  for i in colours:
    (ir, ig, ib, ia) = input;
    (cr, cg, cb, ca) = i;
    dr = ir-cr;
    dg = ig-cg;
    db = ib-cb;
    dist = colour_distance(input, i);
    if dist < least_dist:
      output = i;
      least_dist = dist;
  return output;


def grid(size, pos, im_src):
  (x_size, y_size) = size;
  (x_pos, y_pos) = pos;
  down_sample = im_src.resize((x_size, y_size), PIL.Image.NEAREST);
  down_sample_pixels = down_sample.load();

  for i in range(x_size):
    for j in range(y_size):
      down_sample_pixels[i, j] = get_nearest(down_sample_pixels[i, j]);

  up_sample = down_sample.resize((x_size * 16, y_size * 16), resample = PIL.Image.NEAREST);
  pixels = up_sample.load();

  for y in range(y_size):
    for x in range(x_size):

      for i in range(16):
        pixels[16*x + i, 16*y + 15] = (128, 128, 128, 255);
        pixels[16*x + 15, 16*y + i] = (128, 128, 128, 255);
      
      text_pos = (1 + 16 * x, 1 + 16 * y);
      index = (x_pos + x, y_pos + y);


      colour = get_best_text_colour(pixels[text_pos[0], text_pos[1]]);

      x_str = str(index[0]);
      x_offset = 0;
      for c in x_str:
        digit = int(c);
        width = print(text_pos[0] + x_offset, text_pos[1], digit, pixels, colour);
        x_offset = x_offset + width;

      y_str = str(index[1]);
      y_offset = 0;
      for c in y_str:
        digit = int(c);
        width = print(text_pos[0] + y_offset, text_pos[1] + 6, digit, pixels, colour);
        y_offset = y_offset + width;
  return up_sample;

```
